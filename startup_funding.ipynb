3.cleaning Data

```python
df["IndustryVertical"].value_counts()
```

-------------------------------------------------------------------------------------------------------------------------------

Here from above output we can see that the  vlaue counts on industryvertical column here almost 80% of the data or values we are getting from top three fields i.e Consumer Internet,Technology,eCommerce.So make our data more appropriate for analysis wee are only taking these three category by using isin function and rest of the data of the industryvertical column we are discarding them.

```python
filter = df["IndustryVertical"].isin(["Consumer Internet", "Technology" , "eCommerce"])
df[filter]
```

We are removing the Remarks column from the dataset for analysing the data carefully.

-------------------------------------------------------------------------------------------------------------------------------

```python
New = df.drop(columns=["Remarks"])
New
```

```python
New.columns
```

-------------------------------------------------------------------------------------------------------------------------------

From using the dropna function we are droping out the NaN value from the dataset.

```python
New=New.dropna(subset=["AmountInUSD","CityLocation","IndustryVertical","SubVertical","InvestmentType"])
```

```python
New
```

-------------------------------------------------------------------------------------------------------------------------------

Here in the dataset we have to perform some operation on the AmountInUSD column but the value in the column is in the form of string soo we have to remove all the commas inside that value.

```python
New['AmountInUSD'] = New['AmountInUSD'].str.replace(',', '')
New
```

-----------------------------------------------------------------------------------------------------

Once the data is collected ,cleaned, and processed, it is ready for Analysis.
